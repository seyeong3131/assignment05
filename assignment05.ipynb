```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import signal, misc
from skimage import io, color
from skimage import exposure

## library

file_image = 'cau.jpg'

im_color = io.imread(file_image)
im_gray  = color.rgb2gray(im_color)

# print(np.asarray(im_color))
# print(im_gray)

x_kernel = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
y_kernel = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
emboss_kernel = np.array([[-2,-1,1],[-1,1,1],[0,1,2]])
s_kernel = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
grad_kernel = np.array([[ -3-3j, 0-10j,  +3 -3j], [-10+0j, 0+ 0j, +10 +0j], [ -3+3j, 0+10j,  +3 +3j]])

# X-direction Image
im_convx= signal.convolve2d(im_gray, x_kernel, 'valid')
# Y-direction Image 
im_convy= signal.convolve2d(im_gray, y_kernel, 'valid') 
# Edge detection Image
im_scharr = signal.convolve2d(im_gray, grad_kernel, 'valid')
# Smoothing Image
im_sm = signal.convolve2d(im_gray, s_kernel, 'valid')
# Embossing Image
im_emboss = signal.convolve2d(im_gray, emboss_kernel, 'valid')


p1 = plt.figure(1)
plt.title("Color Image")
plt.imshow(im_color)
plt.axis('off')

p2 = plt.figure(2)
plt.title("Gray Image")
plt.imshow(im_gray, cmap='gray')
plt.axis('off')

p2 = plt.figure(3)
plt.title("Derivative X-direction")
plt.imshow(im_convx, cmap='gray')
plt.axis('off')

p2 = plt.figure(4)
plt.title("Derivative Y-direction")
plt.imshow(im_convy, cmap='gray')
plt.axis('off')


p2 = plt.figure(5)
plt.title("Absolute Gradient")
plt.imshow(np.absolute(im_scharr), cmap='gray')
plt.axis('off')

p5 = plt.figure(6)
plt.title("Smoothing Image")
plt.imshow(im_sm, cmap='gray')
plt.axis('off')

p5 = plt.figure(7)
plt.title("Embossing Image")
plt.imshow(im_emboss, cmap='gray')
plt.axis('off')

plt.show()
```
